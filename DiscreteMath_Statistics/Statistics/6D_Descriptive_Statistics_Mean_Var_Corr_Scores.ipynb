```python
import numpy as np
import matplotlib.pyplot as plt
```

Note: Can use n and (n-1) interchangeably
- n: sample size
- n-1: adjusted sample size

#### Mean
$$\mu = \bar{x} = \frac{\sum x_i}{n}$$ 

#### Variance

$$\sigma^2 = \frac{\sum(x_i-\mu)^2}{n}$$

#### Standard Deviation

$$\sigma = \sqrt{\frac{\sum(x_i -\mu)^2}{n}}$$

```python
x = [1,2,6,4,10]

print("Mean:", np.mean(x))
print("Standard deviation:", np.std(x))
```

## Correlation Coefficient
- High correlation
    + 2 sets of data are strongly linked together

- If
    + corrcoef(x,y) > 0: x,y increase/decrease together
    + corrcoef(x,y) < 0: 1 increase, the other decrease
    + corrcoef(x,y) = 0: x,y not related

- corrcoef() in range [-1,1] 
    + 1 is a perfect positive correlation
    + 0 is no correlation 
        + the values not linked at all
    + -1 is a perfect negative correlation

- **Correlation Is Not Causation!**
    + correlation does not prove one thing causes the other
    + High correlation possibly
        + One event might cause the other
        + The other might cause the first to happen
        + They may be linked by a different reason
        + Or the result could have been random chance

```python
x = [1,2,6,4,2]
y = [3,4,6,5,4]

print(np.corrcoef(x,y))

fig, ax = plt.subplots()
ax.plot(x)
ax.plot(y)
plt.show()
```

```python
x = [1,2,6,4,2]
y = [6,4,2,5,7]

print(np.corrcoef(x,y))

fig, ax = plt.subplots()
ax.plot(x)
ax.plot(y)
plt.show()
```

# Scores

#### Z-score
- Distance above or below the population mean of an observation (unit = 1 std)
- $\sum{z_i} = 0$
$$z_i = \frac{x_i-\mu}{\sigma}$$


```python
from scipy import stats

x = [1,2,6,4,10]
stats.zscore(x)
```

#### R Score - Pearson's Correlation
- Direction(positive/negative) and strength(strong/weak) of a linear correlation
- In range [-1, 1]
    + -1: Strong Negative
    + 0: 
    + 1: Strong Positive

$$r = \frac{\sum{z_xz_y}}{n}$$

- Visualized by the regression of (x,y)

```python
fig, ax = plt.subplots()
x = np.array([1,2,6,4,2])
y = np.array([3,4,6,5,4])

ax.plot(x, y, 'bo')
plt.show()
```
